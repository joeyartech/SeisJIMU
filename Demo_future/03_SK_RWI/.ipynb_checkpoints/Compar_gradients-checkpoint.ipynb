```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n=301; h=10
```

```python
def plot(data,clip=None,title=None):
    
    if clip==None: 
        tmp=np.percentile(np.abs(data),q=99)
        clip=[-tmp,tmp]
    
    plt.imshow(data,vmin=clip[0],vmax=clip[1],aspect='auto')
    plt.colorbar(location='right')
    #plt.xlabel(labels[0]); plt.ylabel(labels[1])
    if title!=None: plt.title(title)

def spec(data):
    full=np.abs(np.fft.fft2(data[:n-1,:n-1]))
    return full[0:(149+1)//2,0:(149+1)//2]
```

```python
#plot JFWI gradients
#where the init rho contains stronger perturbations
plt.figure(figsize=(18,4))

data=np.fromfile('results_JFWI/init_rho-1',dtype='float32').reshape(n,n).T
plt.subplot(1,3,1);plot(data,clip=[0.99,1.01])

data=np.fromfile('results_JFWI/init_rho-5',dtype='float32').reshape(n,n).T
plt.subplot(1,3,2);plot(data,clip=[0.99,1.01])

data=np.fromfile('results_JFWI/init_rho-10',dtype='float32').reshape(n,n).T
plt.subplot(1,3,3);plot(data,clip=[0.99,1.01])

plt.figure(figsize=(18,4))

data=np.fromfile('results_JFWI/cb%gradient_init_rho-1',dtype='float32').reshape(n*6,n).T
data=data[:,n*5:]
plt.subplot(1,3,1);plot(data)

data=np.fromfile('results_JFWI/cb%gradient_init_rho-5',dtype='float32').reshape(n*6,n).T
data=data[:,n*5:]
plt.subplot(1,3,2);plot(data)

data=np.fromfile('results_JFWI/cb%gradient_init_rho-10',dtype='float32').reshape(n*6,n).T
data=data[:,n*5:]
plt.subplot(1,3,3);plot(data)
```

```python
#plot WPI gradient components
plt.figure(figsize=(18,4))

data1=np.fromfile('../03_SK_WPI/results/du_star_a',dtype='float32').reshape(n,n).T
plt.subplot(1,3,1);plot(data1)

data2=np.fromfile('../03_SK_WPI/results/u_star_da',dtype='float32').reshape(n,n).T
plt.subplot(1,3,2);plot(data2)

plt.subplot(1,3,3);plot(data1+data2)
```

```python
#plot the gradients in abs wavenumber

#JFWI
plt.figure(figsize=(18,4))

data=np.fromfile('results_JFWI/cb%gradient_init_rho-1',dtype='float32').reshape(n*6,n).T
data=data[:,n*5:]
plt.subplot(1,3,1); plot(spec(data),clip=[0,8e-8])

data=np.fromfile('results_JFWI/cb%gradient_init_rho-5',dtype='float32').reshape(n*6,n).T
data=data[:,n*5:]
plt.subplot(1,3,2); plot(spec(data),clip=[0,8e-8])

data=np.fromfile('results_JFWI/cb%gradient_init_rho-10',dtype='float32').reshape(n*6,n).T
data=data[:,n*5:]
plt.subplot(1,3,3); plot(spec(data),clip=[0,8e-8])

#WPI
plt.figure(figsize=(18,4))

data1=np.fromfile('../03_SK_WPI/results/du_star_a',dtype='float32').reshape(n,n).T
plt.subplot(1,3,1); plot(spec(data1),clip=[0,6e-13])

data2=np.fromfile('../03_SK_WPI/results/u_star_da',dtype='float32').reshape(n,n).T
plt.subplot(1,3,2); plot(spec(data2),clip=[0,6e-13])

plt.subplot(1,3,3); plot(spec(data1+data2),clip=[0,6e-13])
```
